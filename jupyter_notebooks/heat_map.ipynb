## Create a diluted version of the data frame (fewer data points per postal code):

```python
possible_postal_codes = list(set(symptoms_table.PLZ))
```

```python
symptoms_tables_by_plz = {}
for postal_code in possible_postal_codes:
    symptoms_tables_by_plz[postal_code] = symptoms_table[symptoms_table.PLZ==postal_code]
```

```python
symptoms_tables_by_plz[possible_postal_codes[0]]
```

```python
len(symptoms_tables_by_plz)
```

```python
symptoms_tables_by_plz_half = {}
for plz, data in symptoms_tables_by_plz.items():
    num_rows = data.shape[0]
    num_rows_half = round(num_rows/2)
    symptoms_tables_by_plz_half[plz] = data.iloc[:num_rows_half,:]
```

```python
symptoms_tables_by_plz_half[8192].shape
```

## Take the first 1500 data points of the data frame:

```python
symptoms_table_light = pd.concat(symptoms_tables_by_plz_half)
symptoms_table_light = symptoms_table_light.iloc[:1500,:]
```

```python
symptoms_table_light.head()
```

##  Create an unfiltered heat-map (the map will simply reflect the density of people who provided information):

```python
latitudes = symptoms_table_light['lat']
longitudes = symptoms_table_light['lng']
```

```python
# Store our latitude and longitude


# Creating the location we would like to initialize the focus on. 
# Parameters: Lattitude, Longitude, Zoom
gmap = gmplot.GoogleMapPlotter(latitudes[0], longitudes[0], 10)

# Overlay our datapoints onto the map
gmap.heatmap(latitudes, longitudes)
#gmplot.heatmap_weighted(latitudes, longitudes, fever)

# Generate the heatmap into an HTML file
gmap.draw("my_heatmap_unfiltered.html")
```
