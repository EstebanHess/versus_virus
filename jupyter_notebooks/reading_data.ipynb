## Convert question number 1 into dummy variables:

```python
question_1_possible_answers = list(set(en_quest_table.question_1))
```

```python
question_1_possible_answers
```

```python
question_1_decomp = []
for answer in question_1_possible_answers[1:]:
    #print(answer)
    answer_decomp = answer.split(', ')
    print(answer_decomp)
    for part in answer_decomp:
        if part not in question_1_decomp:
            print(part)
            question_1_decomp.append(part)
```

```python
len(question_1_decomp)
```

```python
check_completeness = []
for answer in question_1_possible_answers[1:]:
    #print(answer)
    answer_decomp = answer.split(', ')
    print(answer_decomp)
    for part in answer_decomp:
        check_bool = part not in question_1_decomp
        check_completeness.append(check_bool)
```

```python
sum(check_completeness)
```

```python
vars_to_dummy_vars = {}
for var_name in question_1_decomp:
    print(var_name)
    vars_to_dummy_vars[var_name] = []
```

```python
vars_to_dummy_vars
```

```python
for answer in en_quest_table.question_1:
    for var_name, data in vars_to_dummy_vars.items():
        #print(answer)
        #print('\n')
        #print(answer)
        #print(var_name)
        if type(answer) is str:
            check_bool = var_name in answer
        else:
            check_bool = np.NaN
        
        #print(check_bool)
        #print(answer)
        #print(var_name)
        vars_to_dummy_vars[var_name].append(check_bool)
```

```python
dummy_vars_question_1 = pd.DataFrame(vars_to_dummy_vars)
dummy_vars_question_1.head()
```

## Encode answers to question 2 by scale:

```python
possible_answers = list(set(en_quest_table.question_2))
possible_answers
```

```python
answers_to_scale = {'Not really': 0, 'Yes, moderatly': 1, 'Mostly yes': 2, 'Yes, mostly': 2,
                   'Yes': 3}
```

```python
new_column = []
for answer in en_quest_table.question_2:
    if type(answer) is str:
        for possible_answer, value in answers_to_scale.items():
            print('\n')
            print(answer)
            print(possible_answer)
            if possible_answer == answer:
                new_column.append(value)
    else:
        new_column.append(np.NaN)
len(new_column)
```

## Encode answers to question 5 by scale:

```python
possible_answers = list(set(en_quest_table.question_5))
possible_answers
```

```python
answers_to_scale = {'A little': 1, 'Much': 2, 'I am an expert': 3}
```

```python
new_column = []
for answer in en_quest_table.question_5:
    if type(answer) is str:
        for possible_answer, value in answers_to_scale.items():
            if possible_answer == answer:
                new_column.append(value)
    else:
        new_column.append(np.NaN)
len(new_column)
```
